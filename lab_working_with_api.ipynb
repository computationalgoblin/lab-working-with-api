## Function 

```python
def get_airport_codes(location):
    url = "https://aerodatabox.p.rapidapi.com/airports/search/term"
    headers = {
        "X-RapidAPI-Host": "aerodatabox.p.rapidapi.com",
        "X-RapidAPI-Key":"9613ca6ca5msh62659872159916fp1a0a2cjsndd9a3e6ff324"  
    }
    query_params = {
        "q": location.lower(),
        "limit": "10"
    }
    response = requests.get(url, headers=headers, params=query_params)
    if response.ok:
        return response.json().get('items', location)
    else:
        print("error:", response.status_code)
        return location


def code_select():
    while True:
        location = input('Insert location: ')
        codes = get_airport_codes(location)
        
        if isinstance(codes, str) or len(set(code['shortName'] for code in codes)) != 1:
            if not codes:
                print('Not Found:', location, '\n')
            else:
                print('Select one of the following locations:', [code['shortName'] for code in codes], '\n')
        else:
            return codes[0]['icao']


def time(from_, to_):
    url = f"https://aerodatabox.p.rapidapi.com/airports/icao/{from_}/distance-time/{to_}"
    resp = requests.get(url, headers=headers)
    return pd.json_normalize(resp.json()).T
```

```python
from_ = code_select()
to_ = code_select()

data = time(from_, to_)
data
```
